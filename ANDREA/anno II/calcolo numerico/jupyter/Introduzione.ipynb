# Funzioni utili per lo svolgimento degli esercizi

## La funzione ```fsolve``` della libreria ```scipy.optimize```

La funzione ```fsolve``` si usa per calcolare una radice di una funzione ```f``` non lineare, il più vicina possibile ad un certo valore ```x0``` specificato come secondo parametro.  
Un'ipotetica chiamata di funzione potrebbe essere la seguente:  
```fsolve(f, -5)```  
che calcolerebbe lo zero di ```f``` più vicino ad ```x = -5```.

````python
import numpy as np
from scipy import optimize

# definiamo la funzione f = cos(x) - x
f = lambda x : np.cos(x) - x

# calcoliamone lo zero più vicino ad x=-2 con fsolve 
result = optimize.fsolve(f, -2)

print("Risultato = ", result)
````

### Situazione particolare usando ```fsolve```
Certe funzioni non hanno radici (non si azzerano mai). In questi casi, ```fsolve``` tende a scrivere in console un messaggio di warning.  
Ad esempio, per ```g(x) = 1/x```:

````python
# g(x) = 1/x
g = lambda x : 1 / x

result = optimize.fsolve(g, 0)

print("Risultato = ", result)
````

Il messaggio "The iteration is not making good progress, as measured by the improvement from the last ten iterations" indica che non vi siano stati significativi cambiamenti nella ricerca di una radice.

## Plotting di un grafico con ```matplotlib```

Per disegnare il grafico di una funzione si usa ```matplotlib.pyplot```, spesso in combinazione con ```numpy```.  
Anzitutto, definiamo un numero di punti da prendere in considerazione durante il plotting del grafico, per poi creare un array ```x``` contenente questi ```n``` punti, mediante la funzione ```linspace``` del modulo ```numpy```:  

````python
# prendo un intervallo [-1, 1]
a = -1; b = 1
# creo un array di n punti tra a e b (di default n = 50)
points = np.linspace(a, b)
````

In seguito, prepariamo lo spazio in cui andare a disegnare il grafico, mediante la property ```figsize``` di ```pyplot```:

````python
import matplotlib.pyplot as plt

# preparo uno spazi largo 12 e alto 6
plt.figure(figsize = (12, 6))
````

Ora prepariamo la funzione ```f``` di cui disegnare il grafico e plottiamo quest'ultimo:

````python
# f(x) = x - cos(x)
f = lambda x : x - np.cos(x)

# calcoliamo f(p), dove p corrisponde ad un punto di points
# e plottiamo il grafico di tale operazione
plt.plot(points, f(points))
````

Nel caso volessimo anche aggiungere un asse al nostro grafico, basterebbe plottare una nuova funzione, passando come punti ```x``` i due estremi del nostro intervallo scelto (quindi ```a``` & ```b```) e come ```y``` due zeri (diciamo quindi di plottare da ```(-1, 0)``` a ```(1, 0)```. Ad esempio:

````python
# plotto da (-1, 0) a (1, 0) una linea 'g' => green (potevo anche usare 'r' => red, etc...)
plt.plot(points, f(points))
plt.plot([a, b], [0, 0], 'g')
````

Ed ecco la funzione ```f(x) = x - cos(x)``` nell'intervallo ```[-1, 1]```!

Nel caso si volesse aggiungere una griglia al grafico, basterebbe inoltre impostare il parametro ```grid``` di ```plt``` a ```true```:

````python
plt.plot(points, f(points))
plt.plot([a, b], [0, 0], 'g')
plt.grid(True)
````
